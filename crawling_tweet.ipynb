```python
import tweepy
import pandas as pd

def crawling(keyword):
  consumer_key = #api-key
  consumer_secret = #api-secret-key
  access_token = #access-token
  access_token_secret = #access-secret-token

  auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
  auth.set_access_token(access_token, access_token_secret)
  api = tweepy.API(auth, wait_on_rate_limit=True)

  search_words = keyword
  new_search = search_words + " -filter:retweets"
  # date_since = "2020-10-10" ##bisa dipake bisa engga
  # date_until = "2021-03-31" ##bisa dipake bisa engga
  
  tweets = tweepy.Cursor(api.search,
                        q=new_search,
                        # since = date_since,
                        # until = date_until,
                        tweet_mode="extended",
                        # count=500,
                        lang="id").items(1000)
  items = []
  tweetId = []
  tgl = []
  idu = []
  id_str = []
  username = []
  name = []
  location = []
  # profile_location = []
  desc = []
  url = []
  # entities = []
  protected = []
  folcount = []
  frcount = []
  listcount = []
  accreat = []
  favcount = []
  utf_off = []
  geoena = []
  verified = []
  statuscount = []
  lang = []
  # statusid = []

  for tweet in tweets:
    items.append(str(tweet.full_text))
    tweetId.append(str(tweet.id))
    tgl.append(tweet.created_at)
    idu.append(str(tweet.user.id))
    id_str.append(str(tweet.user.id_str))
    name.append(tweet.user.name)
    username.append(tweet.user.screen_name)
    location.append(str(tweet.user.location))
    # profile_location.append(str(tweet.user.profile_location))
    desc.append(str(tweet.user.description))
    url.append(str(tweet.user.url))
    # entities.append(str(tweet.user.entities))
    protected.append(str(tweet.user.protected))
    folcount.append(str(tweet.user.followers_count))
    frcount.append(str(tweet.user.friends_count))  
    listcount.append(str(tweet.user.listed_count))
    accreat.append(str(tweet.user.created_at))
    favcount.append(str(tweet.user.favourites_count))
    utf_off.append(str(tweet.user.utc_offset))
    geoena.append(str(tweet.user.geo_enabled))
    verified.append(str(tweet.user.verified))       
    statuscount.append(str(tweet.user.statuses_count))
    lang.append(str(tweet.user.lang))
 
        
  hasil = pd.DataFrame(list(zip(id_str, accreat, desc, location,  url,  protected,  
                                utf_off, geoena, verified, lang, folcount, frcount, listcount, favcount,
                                statuscount,  tweetId, tgl, name, username, items)),
                    columns=['id_str', 'account_created_at', 'description', 'location', 'url', 'protected',  
                            'utf_offset', 'geo_enabled', 'verified', 'lang', 'followers_count', 'friends_count', 
                             'listed_count ',  'favourites_count ', 'status_count', 'tweetid','date', 'name', 'username','tweet'])

  return hasil
```

```python
df1 = crawling('#MosiTidakPercaya')
df1 = df1.drop_duplicates(subset='tweetid', keep="last")
df1.to_excel(r'#MosiTidakPercaya.xlsx')
```
